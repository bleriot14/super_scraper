```python
def exit_function(driver):
    """
    Belirtilen XPath'e sahip bir öğenin sayfada bulunup bulunmadığını kontrol eder.
    
    :param driver: Selenium WebDriver örneği.
    :return: Öğelerden biri bulunursa True, aksi halde False.
    """
    xpath = "//div[@class='_p13n-zg-nav-tree-all_style_zg-browse-item__1rdKf _p13n-zg-nav-tree-all_style_zg-browse-height-large__1z5B8' and .//a]/following-sibling::div[@role='group']"
    try:
        # Öğeyi bulmaya çalışır. Bulunursa devam eder, bulunamazsa NoSuchElementException fırlatır.
        element = driver.find_element(By.XPATH, xpath)
        if element:
            return True
    except NoSuchElementException:
        return False
    return False
```

```python
print(scraper.scrape_menu([(By.XPATH,"//div[@role='group']//div[@role='treeitem']//a")],exit_function))
```
